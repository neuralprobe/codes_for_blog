```python
import os

blog_directory = '/Users/jhoons/Blog/myblog'
draft_directory = '_drafts'
draft_name = '2021-05-23-invent_and_wander.md'

dr = os.path.join(blog_directory,draft_directory,draft_name)
# print(dr)
count = 0
dummyList = ['Jong Hoon Shin', 'Like', 'Write a comment…','Visible:','Delete','Mark as Spoiler']
quote = ''
with open(dr) as fp:
    for line in fp:
        if line.strip()=='FromGoodRead':
            count = 1
            continue
        if count == 1 and line.strip() not in dummyList and not (len(line.strip())<10 and '%' in line.strip()):
            quote += '> '+line.strip()+'\n\n'
        
print(quote)
```
